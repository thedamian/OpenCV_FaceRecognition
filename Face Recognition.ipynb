Import our libraries: OpenCV, face_recognition, numpy, PIL, and matplotlib

```python
import cv2
import face_recognition
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

Now load one of the images. Try `Akaka-Falls.jpg` first, then we will try `c7.jpg` later.

```python
image = face_recognition.load_image_file("DamianAmanda.jpg")
```

Now get the face locations

```python
face_locations = face_recognition.face_locations(image)
```

Let's look at the image

```python
plt.imshow(image)
```

Loop through the detected faces and chop them out

```python
image_array = []
for face in face_locations:
    top,right, bottom, left = face
    face_image = image[top:bottom, left:right]
    f_image = Image.fromarray(face_image)
    image_array.append(f_image)
```

How many faces did it detect?

```python
len(image_array)
```

Let's create an image grid and show all the faces detected.

```python
image_array[1]
```

```python
from mpl_toolkits.axes_grid1 import ImageGrid
import math

numColumns = 2
numRows = math.ceil(len(image_array) / numColumns)
```

```python
fig = plt.figure(figsize=(12.,10.))
grid = ImageGrid(fig,rect=111,nrows_ncols=(numRows,numColumns), axes_pad=0.1)
for axes, img in zip(grid, image_array):
    axes.imshow(img)
plt.show()
```
